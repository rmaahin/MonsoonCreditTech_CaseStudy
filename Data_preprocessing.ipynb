### Feature Selection based on NaN values

I check for the percentage of NaN values for each feature and if the percentage exceeds the threshold of 75%. If the percentage is lower than 5%, then I drop the respective rows instead. I simply drop the column. Otherwise I try to fill up the missing values using differenct methods.

```python
# Percentage of number of NaN values in each column

nan_perc = {}
for column in train_df.columns:
    nan_count = pd.Series(train_df[column]).isna().sum()
    perc = (nan_count/len(train_df))*100
    nan_perc[column] = perc

nan_perc
```

```python
# Create a list of all the columns with more than 75% NaN values

drop_columns = []
for key, value in nan_perc.items():
    if value > 75.0:
        drop_columns.append(key)
        
print('Columns with more than 75% of NaN values are: ', drop_columns)
```

```python
# Dropping the columns

train_df.drop(drop_columns, axis=1, inplace = True)
train_df.head()
```

```python
# Creating a list of all the columns with less than 5% NaN values

drop_rows = []
for key, value in nan_perc.items():
    if value < 5.0 and value > 1.0:
        drop_rows.append(key)
        
print('Columns with less than 5% of NaN values are: ', drop_rows)
```

```python
# Creating a set of all the indices with NaN values for the columns with <5% NaN values

rows = set()
for col in drop_rows:
    for i in range(len(train_df)):
        if math.isnan(train_df.loc[i, col]):
            rows.add(i)
```

```python
# Dropping the rows

train_df.drop(rows, axis=0, inplace=True)
train_df
```

```python
train_df.reset_index(inplace= True, drop=True)
train_df
```

```python
# Creating a list of all the columns with less than 75% and greater than 5% NaN values

fillingna = []
for key, value in nan_perc.items():
    if value < 75.0 and value > 5.0:
        fillingna.append(key)
        
print('Columns with less than 75% and more than 5% of NaN values are: ', fillingna)
```

```python
# Filling NaN values with mean

for col in fillingna:
    train_df[col] = train_df[col].fillna(round(train_df[col].mean(), 1))

train_df
```

```python
train_df.to_csv('preprocessed_data.csv', index=False)
```
