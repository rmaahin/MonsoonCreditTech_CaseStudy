```python
import os
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
test_df = pd.read_csv('Test/X_test.csv')
```

```python
test_df.head()
```

```python
# Percentage of number of NaN values in each column

nan_perc = {}
for column in test_df.columns:
    nan_count = pd.Series(test_df[column]).isna().sum()
    perc = (nan_count/len(test_df))*100
    nan_perc[column] = perc

nan_perc
```

```python
# Create a list of all the columns with more than 75% NaN values

drop_columns = []
for key, value in nan_perc.items():
    if value > 75.0:
        drop_columns.append(key)
        
print('Columns with more than 75% of NaN values are: ', drop_columns)
```

```python
test_df.drop(drop_columns, axis=1, inplace = True)
test_df.head()
```

```python
test_df
```

```python
# Creating a list of all the columns with less than 75% and greater than 5% NaN values

fillingna = []
for key, value in nan_perc.items():
    if value < 75.0 and value>0:
        fillingna.append(key)
        
print('Columns with less than 75% and more than 5% of NaN values are: ', fillingna)
```

```python
# Filling NaN values with mean

for col in fillingna:
    test_df[col] = test_df[col].fillna(round(test_df[col].mean(), 1))

test_df
```

```python
test_df.to_csv('preprocessed_test_data.csv', index=False)
```
